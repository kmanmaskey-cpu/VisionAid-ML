```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
import tensorflow as tf

df = pd.read_csv("Titanic-Dataset.csv")
df['Age'].fillna(df['Age'].median(), inplace=True) #fills missing ages with median age
a = int(input("enter your p class (1,2,3) :"))
b = int(input("enter your age :"))
c = int(input('enter your sex 0=female and 1=male'))
d = float(input("enter your fare :"))

X = df[['Pclass', 'Age', 'Sex', 'Fare']]
y = df['Survived']
model = LogisticRegression(max_iter=1000000)
model.fit(X, y)
user_data = np.array([[a, b, c, d]])
norm_l = tf.keras.layers.Normalization(axis=-1)
norm_l.adapt(user_data)
user_data = norm_l(user_data)
prediction = model.predict(user_data)
print(prediction)
if prediction[0] == 1:
    print("The person survived.")
else:
    print("The person did not survive.")

train_predictions = model.predict(X)
accuracy = np.mean(train_predictions == y)
print(f"Model Accuracy = {accuracy*100:.2f}%")






```
