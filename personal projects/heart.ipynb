```python
import numpy as np
import pandas as pd
import random
from sklearn.linear_model import LogisticRegression

df = pd.read_csv("Heart_Disease_Prediction.csv")
a = int(input("Enter Thallium (6,3,7): "))
b = float(input("Enter st depression induced by exercise relative to rest: "))
c = int(input("Enter exercise induced angina (1 = yes; 0 = no): "))
d = int(input("Enter chest pain type (1-4): "))
e = int(input("Enter Number of vessels fluro: "))

X = df[['Thallium', 'ST depression', 'Number of vessels fluro', 'Chest pain type', 'Exercise angina']]
y = df['Heart Disease'].map({'Absence': 0, 'Presence': 1})
model = LogisticRegression(max_iter=1000000)
model.fit(X, y)
user_data = np.array([[a, b, e, d, c]])
prediction = model.predict(user_data)
print(prediction)
if prediction[0] == 1:
    print("The person has heart disease.")
else:
    print("The person does not have heart disease.")

     




```
