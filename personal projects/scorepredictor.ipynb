```python
import numpy as np
import pandas as pd
import random
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

df = pd.read_csv("Exam_Score_Prediction.csv")

a= float(input("Enter your age: "))
c= float(input("Enter your class attendance: "))
d= input("Enter your internet access yes or no: ")
e= float(input("Enter your sleep hours: "))
f= input("Enter your sleep quality poor average or good: ")
g= input("Enter your study method coaching,online videos,mixed,self-study,group study")
i=input("Enter your facility rating low medium high: ")
h= input("Enter your exam difficulty easy moderate hard: ")
m=input("enter you course diploma,bca,b.sc,ba,b.com,b.tech,bba ")
n=input("enter your gender male ,female or other:")

age = df['age']
class_attendance = df['class_attendance']
internet_access = df['internet_access']
sleep_hours = df['sleep_hours']
sleep_quality = df['sleep_quality']
study_method = df['study_method']
facility_rating = df['facility_rating']
exam_difficulty = df['exam_difficulty']

categorical_cols = ['gender', 'course', 'internet_access', 
                    'sleep_quality', 'study_method', 'facility_rating',
                    'exam_difficulty']

df_encoded = pd.get_dummies(df, columns=categorical_cols)

X_train = df_encoded[df_encoded.columns.difference(['exam_score'])]
y_train = df_encoded['exam_score']
model = LinearRegression()
model.fit(X_train, y_train)

user_data = {
    'age': a,
    'class_attendance': c,
    'internet_access': d,  # 'yes' or 'no'
    'sleep_hours': e,
    'sleep_quality': f,    # 'poor', 'average', 'good'
    'study_method': g,     # 'coaching', 'online videos', etc.
    'exam_difficulty': h,  # 'easy', 'moderate', 'hard'

    'gender': n,  # You missed this!
    'course': m,  # You missed this!
    'facility_rating': i,  # You missed this!
}
user_df = pd.DataFrame([user_data])
user_encoded = pd.get_dummies(user_df, columns=categorical_cols)
missing_cols = set(X_train.columns) - set(user_encoded.columns)
for col in missing_cols:
    user_encoded[col] = 0
user_encoded = user_encoded[X_train.columns]

predicted_score = model.predict(user_encoded)[0]
print(f"Predicted Exam Score: {predicted_score:.2f}")

u = mean_squared_error(y_train, model.predict(X_train))
print(f"Training Mean Squared Error: {u:.2f}")      
```
