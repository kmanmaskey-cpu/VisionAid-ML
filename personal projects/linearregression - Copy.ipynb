```python
import numpy as np
import pandas as pd

z = pd.read_csv("life.csv")
y_train = z['Life expectancy']
x_train = z[["Adult Mortality","under-five deaths","Measles","HIV/AIDS","Income composition of resources","Schooling","BMI"]]

def normalize_data(data):
    return (data - data.min()) / (data.max() - data.min())

def cost_linear(x_train,y_train,weight,bias):
    m = len(y_train)
    y_pred = np.dot(x_train,weight)+bias
    cost = (1/(2*m))*np.sum(np.square(y_pred - y_train  ))
    return cost
def linear_regression(x_train,y_train,learning_rate,iterations):
    weight = np.zeros(x_train.shape[1])
    bias = 0
    m = len(y_train)
    for i in range(iterations):
        y_pred = np.dot(x_train,weight)+bias
        gw = (1/m)*np.dot(x_train.T,y_pred-y_train)
        gb = (1/m)*np.sum(y_pred-y_train)
        weight = weight - learning_rate*gw
        bias = bias - learning_rate*gb
    return weight,bias

weight,bias = linear_regression(x_train,y_train,0.0001,1000)

a = float(input('enter Adult Mortality '))
b = float(input('enter under-five deaths '))
c = float(input('enter Measles'))
d = float(input('HIV/AIDS'))
e = float(input('Income composition of resources'))
f = float(input('Schooling'))
g = float(input('BMI'))

o = np.array([[a,b,c,d,e,f,g]])
o = normalize_data(o)
pred = np.dot(o,weight)+bias
print('your life expectancy is ', pred)






    
```
