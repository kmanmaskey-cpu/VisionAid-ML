```python
import numpy as np
import matplotlib.pyplot as plt

# If in Jupyter, keep this line
%matplotlib widget

# Create the missing functions locally
def plt_logistic_cost(y=None, y_hat=None):
    fig, ax = plt.subplots(figsize=(6, 4))
    z = np.linspace(-10, 10, 100)
    sigmoid = 1 / (1 + np.exp(-z))
    ax.plot(z, sigmoid, 'b-')
    ax.set_title('Logistic Sigmoid Function')
    ax.set_xlabel('z')
    ax.set_ylabel('(z)')
    return fig

def plt_two_logistic_loss_curves():
    fig, ax = plt.subplots(figsize=(6, 4))
    x = np.linspace(-5, 5, 100)
    ax.plot(x, 1/(1+np.exp(-x)), 'b-', label='Curve 1')
    ax.plot(x, 1/(1+np.exp(-2*x)), 'r-', label='Curve 2')
    ax.legend()
    ax.set_title('Two Logistic Curves')
    return fig

def plt_simple_example(x, y):
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.scatter(x, y)
    ax.set_title('Simple Example')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    return fig

def soup_bowl():
    fig, ax = plt.subplots(figsize=(6, 4))
    x = np.linspace(-5, 5, 100)
    ax.plot(x, x**2, 'b-')
    ax.set_title('Cost Function (Soup Bowl)')
    ax.set_xlabel('Parameter')
    ax.set_ylabel('Cost')
    return fig

def plt_logistic_squared_error():
    fig, ax = plt.subplots(figsize=(6, 4))
    x = np.linspace(-2, 2, 100)
    sig = 1/(1+np.exp(-x))
    ax.plot(x, -np.log(sig), 'b-', label='Logistic Loss')
    ax.plot(x, (sig-1)**2, 'r-', label='Squared Error')
    ax.legend()
    ax.set_title('Loss Functions')
    ax.set_xlabel('z')
    ax.set_ylabel('Loss')
    return fig

# Create data
x_train = np.array([0., 1, 2, 3, 4, 5], dtype=np.longdouble)
y_train = np.array([0,  0, 0, 1, 1, 1], dtype=np.longdouble)

# Call functions CORRECTLY (without parameters for those that don't need them)
print("Creating visualizations...")

# This works - plt_simple_example expects x, y
fig1 = plt_simple_example(x_train, y_train)

plt.close('all')

# FIXED: plt_logistic_squared_error doesn't take parameters
fig2 = plt_logistic_squared_error()  # NOT: plt_logistic_squared_error(x_train,y_train)

fig3 = plt_two_logistic_loss_curves()

plt.close('all')

# FIXED: plt_logistic_cost doesn't take parameters  
fig4 = plt_logistic_cost()  # NOT: plt_logistic_cost(x_train,y_train)

# Optional: Also show soup bowl
fig5 = soup_bowl()
print(fig4 )
plt.show()
print("All plots created!")
```
