```python
import numpy as np
import matplotlib.pyplot as plt

# Create your own versions of missing imports
dlc = {
    'dlblue': '#0096ff',
    'dlorange': '#FF9300', 
    'dldarkred': '#C00000',
    'dlmagenta': '#FF40FF',
   
}

def plot_data(x, y, ax=None, title="", xlabel="", ylabel=""):
    """Simple data plotting function"""
    if ax is None:
        fig, ax = plt.subplots()
    # Separate positive (y=1) and negative (y=0) classes
    pos = y == 1
    neg = y == 0
    ax.scatter(x[pos, 0], x[pos, 1], marker='x', s=80, c='red', label="y=1")
    ax.scatter(x[neg, 0], x[neg, 1], marker='o', s=100, label="y=0", 
               facecolors='none', edgecolors=dlc["dlblue"], lw=3)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    return ax

# Simple onclick function for demonstration
def plt_one_addpt_onclick(ax=None):
    if ax is None:
        fig, ax = plt.subplots()
    ax.set_title("Click to add points")
    return ax

print("✓ All imports fixed!")

# Create training data
x_train = np.array([0., 1, 2, 3, 4, 5])
y_train = np.array([0,  0, 0, 1, 1, 1])
X_train2 = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train2 = np.array([0, 0, 0, 1, 1, 1])

# Separate positive and negative classes for first plot
pos = y_train == 1
neg = y_train == 0

# Create figure with 2 subplots
fig, ax = plt.subplots(1, 2, figsize=(8, 3))

# Plot 1: One variable (x vs y)
ax[0].scatter(x_train[pos], y_train[pos], marker='x', s=80, c='red', label="y=1")
ax[0].scatter(x_train[neg], y_train[neg], marker='o', s=100, label="y=0", 
              facecolors='none', edgecolors=dlc["dlblue"], lw=3)
ax[0].set_ylim(-0.08, 1.1)
ax[0].set_ylabel('y', fontsize=12)
ax[0].set_xlabel('x', fontsize=12)
ax[0].set_title('One variable plot')
ax[0].legend()

# Plot 2: Two variables (x₀ vs x₁)
plot_data(X_train2, y_train2, ax[1])  # Now includes labels for legend
ax[1].axis([0, 4, 0, 4])
ax[1].set_ylabel('$x_1$', fontsize=12)
ax[1].set_xlabel('$x_0$', fontsize=12)
ax[1].set_title('Two variable plot')
ax[1].legend()  # Now has something to show!
plt.tight_layout()
plt.show()
# Create a simple plot using the original function
w_in = np.zeros((1,))
b_in = 0
plt.close('all') 

# Create a simple plot (not interactive)
fig, ax = plt.subplots(figsize=(8, 4))
ax = plt_one_addpt_onclick(ax)  # Just pass the axis
plt.show()
```
